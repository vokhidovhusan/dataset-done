```python
with open(file_name[:-4]+'.txt') as f:
    img = plt.imread(file_name)
    fig, ax = plt.subplots(1)
    fig.set_size_inches(40, 10.5)
    height, width, _ = img.shape
    ax.set_aspect('equal')
    ax.imshow(img)
    lines = f.readlines()

    for line in lines:
        line = line.split()        
        print(line)
        x = float(line[1])
        y = float(line[2])
        w = float(line[3])
        h = float(line[4])
        x1 = x - w/2
        x2 = x + w/2
        y1 = y - h/2
        y2 = y + h/2
        rect = patches.Rectangle((x1 * width, y1 * height), w * width, h * height, linewidth=1, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

    plt.show()
```

```python
import json
ROOT_DIR = 'D:\lightvision\projects\pedestrian_detection\dataset\VISUAL_GENOME\Version_1_2'
JSON_FILE = 'objects.json'
annot_path = os.path.join(ROOT_DIR,'annotations', JSON_FILE)
img_path = os.path.join(ROOT_DIR, 'images')


with open(annot_path) as f:
    x = []
    y = []
    data = json.load(f)
    for d in data:
        image_id = d['image_id']
        if not image_id == index:
            continue
        file_name = os.path.join(ROOT_DIR, 'images', str(image_id) + '.jpg')
        img = plt.imread(file_name)
        fig, ax = plt.subplots(1)
        fig.set_size_inches(40, 10.5)
        height, width, _ = img.shape
        ax.set_aspect('equal')
        ax.imshow(img)
        objects = d['objects']
        
        for object in objects:
            if len(object['names']) > 1:
                print(object['names'])
            for name in object['names']: 
                if not name in label_list:
                    continue
                print(name)
                x = object['x']
                y = object['y']
                w = object['w']
                h = object['h']
                x1 = x - w/2
                x2 = x + w/2
                y1 = y - h/2
                y2 = y + h/2
                rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
                ax.add_patch(rect)
                ax.text(x, y, name) # this will draw text with Blackcolor and 16 size
                # plt.text(x, y, s, bbox=dict(fill=False, edgecolor='red', linewidth=2))

        plt.show()
```
